```python
import torch
import torchvision
from torch import nn
from torch.nn import Conv2d
from torch.utils.tensorboard import SummaryWriter 
```

```python
from torch.utils.data import DataLoader

dataset = torchvision.datasets.CIFAR10(root='data', train=True, download=True, transform=torchvision.transforms.ToTensor())

dataloder = DataLoader(dataset, batch_size=64)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.conv1 = Conv2d(in_channels=3, out_channels=6, kernel_size=3, padding=0)
        self.maxpool = nn.MaxPool2d(kernel_size=2, stride=2)
    def forward(self, x):
        x = self.conv1(x)
        return x

model = Model()
tensorboard = SummaryWriter('log')

setp = 0
for data in dataloder:
    images, labels = data
    outputs = model(images)
    print(images.shape)
    print(outputs.shape)
    # torch.Size([6, 30, 30]) torch.Size([3, 32, 32])->[xx,3,30,30]

    tensorboard.add_image('input', images, setp, dataformats="NCHW")
    outputs = torch.reshape(outputs,(-1,3,30,30) )

    tensorboard.add_image('output', outputs, setp, dataformats="NCHW")
    setp = setp + 1
tensorboard.close()
```
